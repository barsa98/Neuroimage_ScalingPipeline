# MR Subtraction Maps between T1 and T2

```python
#################Scaling factor betweeen T1GD AND T1

abs_array = []
fig = plt.figure(figsize= (20, 60))
for i in range(0,155):
    sum_absolute_diff = (np.abs(T1_sub1[:,:,i] - T2_sub1[:,:,i]))
    abs_array.append(sum_absolute_diff)
    ax = fig.add_subplot(32,5, i+1)
    ax.imshow(ndi.rotate(sum_absolute_diff, 270), cmap = "gray")
    ax.set_title(i)
    plt.axis('off')
```
